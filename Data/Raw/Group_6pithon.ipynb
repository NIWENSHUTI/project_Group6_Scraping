```python
from bs4 import BeautifulSoup as bs4
import os

# Step 1: Define file paths
file_paths = [
    "Zimbabwe.txt"
]

# Adjust this to your actual directory
base_path = r"C:\Users\ADELINE PC\Downloads\ESPNCricketInfo"

for file_name in file_paths:
    full_path = os.path.join(base_path, file_name)
    
    # Check if file exists before opening
    if os.path.exists(full_path):
        with open(full_path, "r", encoding="utf-8") as file:
            saved_html = file.read()
        
 # Now parse it with BeautifulSoup
soup = bs4(saved_html, "html.parser")
        
soup
```

```python
# Find all divs with class 'ds-w-full ds-bg-fill-content-prime ds-overflow-hidden ds-rounded-xl ds-border-line ds-mb-4'
ds_w_full = soup.find_all('div', class_='ds-w-full ds-bg-fill-content-prime ds-overflow-hidden ds-rounded-xl ds-border-line ds-mb-4')

# Find all divs with class 'ds-overflow-x-auto ds-scrollbar-hide'
ds_overflow_x_auto_ds_scrollbar_hide= soup.find_all('div', class_='ds-overflow-x-auto ds-scrollbar-hide')

# Combine the two lists
combined_ds = ds_w_full + ds_overflow_x_auto_ds_scrollbar_hide

combined_ds
```

```python
import pandas as pd

# Step 1: Combine the two sets of divs
ds_w_full = soup.find_all('div', class_='ds-w-full ds-bg-fill-content-prime ds-overflow-hidden ds-rounded-xl ds-border-line ds-mb-4')
ds_overflow_x_auto = soup.find_all('div', class_='ds-overflow-x-auto ds-scrollbar-hide')
combined_ds = ds_w_full + ds_overflow_x_auto

# Step 2: Look for <table> tags inside each div
dataframes = []

for section in combined_ds:
    table = section.find('table')
    if table:
        try:
            df = pd.read_html(str(table))[0]  # Only grab the first table if multiple
            dataframes.append(df)
        except ValueError:
            pass  # Ignore sections that don't parse into tables

# Step 3: Combine and export
if dataframes:
    final_df = pd.concat(dataframes, ignore_index=True)
    final_df.to_excel('cricket_data_20.xlsx', index=False)
    print(" Excel file saved as 'cricket_data_20.xlsx'")
else:
    print(" No tables found in the combined divs.")
```
