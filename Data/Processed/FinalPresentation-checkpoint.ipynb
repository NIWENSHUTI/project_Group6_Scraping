## This is the process we passed through to scrap our data in all countries and this is our dataset

##### we first scraped data from the website https://www.espncricinfo.com/records/trophy/world-cup-12

```python
from bs4 import BeautifulSoup as bs4
import os

# Step 1: Define file paths
file_paths = [
    "Afghanistan.txt"
]
# Adjust this to your actual directory
base_path = (r"C:\Users\USER\Downloads\ESPNCricketInfo")

for file_name in file_paths:
    full_path = os.path.join(base_path, file_name)
    
    # Check if file exists before opening
    if os.path.exists(full_path):
        with open(full_path, "r", encoding="utf-8") as file:
            saved_html = file.read()
        
 # Now parse it with BeautifulSoup
soup = bs4(saved_html, "html.parser")
        
soup
```

##### Finding all divs

```python
# Find all divs with class 'ds-grow'
ds_grow = soup.find_all('div', class_='ds-grow')

# Find all divs with class 'ds-popper-wrapper'
ds_popper_wrapper = soup.find_all('div', class_='ds-popper-wrapper')

# Combine the two lists
combined_wrappers = ds_grow + ds_popper_wrapper

combined_wrappers
```

## Saving data frame

```python
import pandas as pd

dataframes = []
ds_grow = soup.find_all('div', class_='ds-grow')

# Find all divs with class 'ds-popper-wrapper'
ds_popper_wrapper = soup.find_all('div', class_='ds-popper-wrapper')

# Combine the two lists
combined_wrappers = ds_grow + ds_popper_wrapper

combined_wrappers

for section in combined_wrappers:
    table = section.find('table')
    if table:
        try:
            df = pd.read_html(str(table))[0]  # Only grab the first table if multiple
            dataframes.append(df)
        except ValueError:
            pass  # Ignore sections that don't parse into tables

# Step 3: Combine and export
if dataframes:
    final_df = pd.concat(dataframes, ignore_index=True)
    final_df.to_excel('cricket_data.xlsx', index=False)
    print(" Excel file saved as 'cricket_data.xlsx'")
else:
    print(" No tables found in the combined divs.")
```

# Data analysis and visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV
df = pd.read_csv(r"C:\Users\USER\Desktop\Updated_Batting_Averages.csv")

# Check column names
print(df.columns)

# Clean and prepare the DataFrame
df_cleaned = df.copy()

# Ensure key numeric columns are in correct format
df_cleaned['Ave'] = pd.to_numeric(df_cleaned['Ave'], errors='coerce')
df_cleaned['SR'] = pd.to_numeric(df_cleaned['SR'], errors='coerce')  # Strike Rate
df_cleaned['Rookie Year'] = pd.to_numeric(df_cleaned['Rookie Year'], errors='coerce')

# === 1. Distribution of Batting Averages ===
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Ave'].dropna(), bins=20, kde=True)
plt.title('Distribution of Player Batting Averages')
plt.xlabel('Batting Average')
plt.ylabel('Number of Players')
plt.grid(True)
plt.tight_layout()
plt.show()



```

```python
# === 2. Top 10 Highest Career Batting Averages ===
top_players = df_cleaned.sort_values(by='Ave', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='Ave', y='Player', data=top_players, hue='Country', dodge=False)
plt.xlabel('Batting Average')
plt.title('Top 10 Players by Batting Average')
plt.legend(title='Country', loc='lower right')
plt.tight_layout()
plt.show()

```

```python
# === 3. Average Batting Strike Rate by Country ===
# Group by country and calculate mean strike rate
avg_strike_rate = (
    df_cleaned.groupby('Country')['SR']
    .mean()
    .dropna()
    .sort_values(ascending=False)
    .reset_index()
)
# Bar chart for avarage batting strike
plt.figure(figsize=(14, 6))
sns.barplot(x='Country', y='SR', data=avg_strike_rate, palette='viridis')
plt.xticks(rotation=45)
plt.ylabel('Average Strike Rate')
plt.xlabel('Country')
plt.title('Average Batting Strike Rate by Country')
plt.tight_layout()
plt.show()

```

```python
# === 4. Rookie Year Trends ===
rookie_counts = df_cleaned['Rookie Year'].dropna().value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x=rookie_counts.index, y=rookie_counts.values, marker='o')
plt.title('Number of New Players per Rookie Year')
plt.xlabel('Rookie Year')
plt.ylabel('Number of Players')
plt.grid(True)
plt.tight_layout()
plt.show()
```
